```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt 

gpus = tf.config.experimental.list_physical_devices(device_type='GPU') #本地需要这样操作
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu ,True)
```

```python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.fashion_mnist.load_data()
x_train=x_train/255
x_test=x_test/255
```

```python
BATCH_SIZE=32
train_count=len(x_train)
dataset_train=tf.data.Dataset.from_tensor_slices((x_train,y_train)).shuffle(train_count).repeat().batch(BATCH_SIZE)
test_count=len(x_test)
dataset_test=tf.data.Dataset.from_tensor_slices((x_test,y_test)).repeat().batch(BATCH_SIZE)
```

```python
keras_input=tf.keras.Input(shape=(28,28))
x=tf.keras.layers.Flatten()(keras_input)
x=tf.keras.layers.Dense(32,activation='relu')(x)
x=tf.keras.layers.Dropout(0.5)(x)
x=tf.keras.layers.Dense(64,activation='relu')(x)
keras_output=tf.keras.layers.Dense(10,activation='softmax')(x)
model=tf.keras.Model(inputs=keras_input,outputs=keras_output)
```

```python
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.sparse_categorical_crossentropy,
    metrics=['acc']
)

```

```python
model.fit(
    x_train,
    y_train,
    epochs=10,
    steps_per_epoch=train_count//BATCH_SIZE,
    validation_data=dataset_test,
    validation_steps=test_count//BATCH_SIZE
)
```
